# Mini Project 2: Sales Data Analysis using Pandas

This notebook demonstrates loading, cleaning, transforming, and aggregating a sales dataset using Python and Pandas.


```python
import pandas as pd
```

```python
df = pd.read_csv("sales_data.csv")
df
```

```python
df.isnull().sum()
```

```python
df["Sales"] = df["Sales"].fillna(df["Sales"].mean())
df["Quantity"] = df["Quantity"].fillna(0)
df["Date"] = pd.to_datetime(df["Date"])
```

```python
df = df.rename(columns={"Order_ID": "OrderID"})
```

```python
sales_by_category = df.groupby("Category")["Sales"].sum().reset_index()
sales_by_category
```

```python
sales_by_region = df.groupby("Region")["Sales"].sum().reset_index()
sales_by_region
```

```python
sales_by_category.sort_values(by="Sales", ascending=False)
```

```python
df.to_csv("cleaned_sales_data.csv", index=False)
sales_by_category.to_csv("sales_by_category.csv", index=False)
sales_by_region.to_csv("sales_by_region.csv", index=False)
```
